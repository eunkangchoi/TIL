```python

from bs4 import BeautifulSoup
from selenium import webdriver
import time

# driver = webdriver.Firefox()
driver= webdriver.Chrome() #크롬엔진을 계속 업데이트해줘야함.
driver.get('http://matstar.sbs.co.kr/program.html?programs=S01_V0000305532')
time.sleep(1)

html = driver.page_source #BeautifulSoup에게 페이지를 넘긴다.
soup = BeautifulSoup(html)
prodList = soup.select('ul.restaurant_list > div')

print(len(prodList)) #17개


results=[]
for prod in prodList:
    try:
        img_url= prod.select_one('div > li.rl_cont > div > a > div.box_module_image_w > img.box_module_image')
        name= prod.select_one('div > li.rl_cont > div > a > div.box_module_cont > div > strong')
        menu=prod.select_one('div > li.rl_cont > div > a > div.box_module_cont > div > div > div.mil_inner_kind > span.il_text')
        address=prod.select_one('div > li.rl_cont > div > a > div.box_module_cont > div > div > div.mil_inner_spot > span.il_text')
        data={
            'img_url':img_url['src'],
            'name': name.text,
            'menu': menu.text,
            'addr': address.text
        }
        results.append(data)
    except Exception as e:
        print('--- Exception')
        print(prod)
        print(e)
        print('---')
driver.close()

print('--- Results')
for r in results:
    print(r)
print('---')

```

# 생생정보통 크롤링

- https://menutong.com/

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
html= requests.get('https://menutong.com/', headers=headers)

soup=BeautifulSoup(html.text, 'html.parser')

# 생생정보통 식당 데이터
restaurants= soup.select('.widget-box > .basic-post-gallery > .post-row')

#검색결과 json형태로
results=[]

for restaurant in restaurants:
    # 이미지 url
    img_url=restaurant.select_one('.post-list > .post-image > a > div.img-wrap > div.img-item > img')['src']
    
    # 가게이름
    name=restaurant.select_one('.post-content > div:nth-child(2) > font > b').text
    
    # 메뉴이름
    menu_kind=restaurant.select_one('.post-content > div:nth-child(3) > font').text
    
    # 가게주소
    address=restaurant.select_one('.post-content > div:nth-child(4)').text
    
    result={
        'img_url': img_url,
        'name': name,
        'menu': menu_kind,
        'addr':address
    }
    
    results.append(result)
    

results
```
