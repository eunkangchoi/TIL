> ## 네이버호텔 샘플

```python
from bs4 import BeautifulSoup
from selenium import webdriver

driver = webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
url='https://hotel.naver.com/hotels/list?destination=place:Seoul&destination_kor=%EC%84%9C%EC%9A%B8'

driver.get(url)
req= driver.page_source
soup= BeautifulSoup(req, 'html.parser')
soup
```

## 호텔이름 

- 호텔이름: strong.hotel_name_ko

```python
hotel_names= soup.select('strong.hotel_name_ko')
hotel_names
```

```python
hotel_names= []
for name in soup.select('strong.hotel_name_ko'):
    hotel_names.append(name.text)
hotel_names
```

<br>

## 호텔등급

- 호텔등급: span.grade

```python
hotel_grade=soup.select('span.grade')
hotel_grade
```

```python
hotel_grades=[]
for grade in soup.select('span.grade'):
    hotel_grades.append(int(grade.text[0]))
hotel_grades
```

<br>

## 호텔 1박 최저가

```python
hotel_low_prices=soup.select('em.ng-binding')
prices=[]
for price_string in hotel_low_prices:
    prices.append(int(price_string.text.replace(',','')))
    
prices
```
